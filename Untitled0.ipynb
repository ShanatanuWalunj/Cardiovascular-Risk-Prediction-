**Recomendation Approaches:**

Collaborative Filtering

In Collaborative Filtering we make predictions of the taste or the interests of the user based on the other similar users behaviour (which is called collaborative).The simple understanding that if a person P1 have similar opinion whith person P2, then if person P1 is reading Harry Potter books then person P2 may also read Harry Potter books

Collaborative Filtering has mainly two implimentation strategies

* Model Based
* Memory Based

We have used the following approaches

Memory Based

* User_based Collaborative Filtering

In this technique we try to predict on basis of what user may like based on the ratings that they had given to the items. In simple words here we try to indentify the neghbouring users on the basis of similarity of active users and the scoring of the items is calculated on the basis of those neighbour users

* Item_based Collaborative Filltering

In item based collaborative filtering we try to find similar items based on the ratings that they had given to the items which users
has liked or interected with. It suggests an item based on items the user has previously consumed. It looks for the items the users has consumed then it finds other items similar to consumed items and recommends accordingly

Model based Collaborative Filtering Approaches

* Singular value decomposition(SVD)

It is a popular method in field of Data Science and Machine Learning. It is a classical method from the Linear Algebra concept. It is popular because of its properties and its use in Recommendation System. We can use this in recommending Books, movies etc.

SVD is a method from linear algebra that the has been generally used as a dimensionality reduction technique in ML problems.It is a matrix Factorization technique which tries to reduce the number of feature or attributes by reducing the space dimension from N dimension to SVD is used as a collaborative filtering techique in recommendation technique domain. It consist of a matrix structure where the rows define the users and columns defines the items.

* Non-Negative Matrix Factorization(NMF):

Our goal in NMF is to approximate the users-item V matrix by the dot product of two arrays W and H. Dimensions of the arrays are defined by dimensions of V and number of components we set to the algorithm if V has n rows and m columns and we want to decompose it to k components,then W has n rows, and k columns and H has k rows and m columns. This is actually matrix factorization part of the algorithm. The Non-negative part refers to V,W, and H - all the values have to be equal or greater that zero, i.e., non-negative

It is type of recommendation system which depends upon the data which we get from the coustomers, based on the data a user profile is generated

## **2. Handling Outliers**

```python
fig = plt.figure(figsize=(6,4))
sns.boxplot(x='Age',data=users_df)
```

```python
# Calculate the IQR for the 'Age' column
Q1 = users_df['Age'].quantile(0.25)
Q3 = users_df['Age'].quantile(0.75)
IQR = Q3-Q1

#Define bounds for outliers
lower_bounds = Q1-1.5*IQR
Upper_bounds = Q3+1.5*IQR

#Identify outliers
outliers = users_df[(users_df['Age']<lower_bounds) | (users_df['Age']>Upper_bounds)]

users_df['Age'] = np.where(users_df["Age"] > Upper_bounds, Upper_bounds,users_df['Age'])
users_df['Age'] = np.where(users_df['Age']< lower_bounds, lower_bounds, users_df['Age'])

#Dispalying the number of outliers and some statistics
num_outliers = outliers.shape[0]
outliers_Age_stats = outliers['Age'].describe()
num_outliers, outliers_Age_stats
```

```python
outliers.sort_values('Age', ascending=False)
```

```python
fig =plt.figure(figsize=(6,4))
sns.boxplot(x='Age',data=users_df)
```

*   We can notice that the outliers has been handeled smoothly and now we dont have any outliers.

**What all outlier treatment techniques have you used and why did you use those techniques?**
*   We used IQR method to fill the right side outliers with maximum value calculated from IQR and Left side outliers with the minimum value calculated from IQR.

### Now, we concider only that books which have minimum 250 rating.

```python
# merge books_df and rating_df dataset on 'ISBN' column
ratings_with_names =  ratings_df.merge(books_df, on='ISBN')
```

```python
ratings_with_names
```

```python
# Now find that names which have rating greater than 250
num_rating_df =  ratings_with_names.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
# Now find that names which have 'Average' rating greater than 250
avg_rating_df =  ratings_with_names.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
# merge above two dataset into popular_df
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
# Now find that Books which have ratings greater than 250
popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False)
```

```python
# Now we have top 50 Books which have high rating
popular_df = popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
popular_df
```

```python
# Now we have this top 50 books auther name, publisher and image
# So we can merge it with Books_df dataset
popular_df.merge(books_df, on='Book-Title').drop_duplicates('Book-Title').head()
```

```python
# Now we have only few columns like 'Book-Title', 'Book-Rating','Avg-Rating','Book-Author' and 'Image'
popular_df = popular_df.merge(books_df, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]

```

# **Collaborative Filtering Based Recommender System.**

We have only those users who give atleast 200 ratings.

```python
m = ratings_with_names.groupby('User-ID').count()['Book-Rating'] > 200
educated_users_df = m[m].index
```

```python
# Now find this in 'ratings_with_names' dataset
filtered_rating = ratings_with_names[ratings_with_names['User-ID'].isin(educated_users_df)]
filtered_rating.head()
```

### Now filter on the basis of 'Book-title'.

```python
 y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
 famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings.head(5)
```

```python
# Check duplicate once again in 'final_rating dataset'
final_ratings.duplicated().sum()
```

```python
# Now its time for pivot the table.
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

### Let's fill 0 in place of NaN.

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
# Now its time to find Euclidean distance between this vector
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# find cosine _similarity score with other rows
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
  #index_fitch
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

  for i in similar_items:
    print(pt.index[i[0]])

```

```python
recommend('Message in a Bottle')
```

```python
def recommend(book_name):
  #index_fitch
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

  data =[]

  for i in similar_items:
    item=[]
    temp_df = books_df[books_df['Book-Title']==pt.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

    data.append(item)

  return data
```
